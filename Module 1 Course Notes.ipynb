## Functions, split()

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

# Return Dr. Lastname

def split_name(person):
    title = person.split()[0]
    surname = person.split()[-1]
    return title + " " + surname
    #or return '{} {}'.format(title, surname)
    #or return " ".join([title, surname])

list(map(split_name, people))
```

```python
for person in people:
    print(split_name(person))
```

```python
list(map(split_name, people))
#option 1
for person in people:
    print(split_name(person) == (lambda x:x.split()[0] + " " + x.split()[-1]))

#option 2
list(map(split_name, people)) == list(map(lambda person: person.split()[0] + " " + person.split()[-1], people))
```

## List comprehension

```python
#list comprehension
my_list = [number for number in range(0, 10) if number % 2 == 0]
my_list
```

```python
def times_tables():
    lst = []
    for i in range(10):
        for j in range (10):
            lst.append(i*j)
    return lst

```

```python
#list comprehension method
times_tables() == [i*j for i in range(10) for j in range(10)]
```
